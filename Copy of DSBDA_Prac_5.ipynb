```python
# Data Correction
```

```python
def remove_outliers(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    threshold = 1.5 * IQR
    outlier_mask = (column < Q1 - threshold) | (column > Q3 + threshold)
    return column[~outlier_mask]
```

```python
heart_name = ['restecg','slp']
for col in heart_name:
    heart[col] = remove_outliers(heart[col])
```

```python
# Heart Data splitting
from sklearn.model_selection import train_test_split

x = heart[['trtbps','fbs','exng','oldpeak']]
y = heart.output
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
# Air Data splitting
from sklearn.model_selection import train_test_split

a = air[['PM10','NO2','NH3']]
b = air.AQI
a_train,a_test,b_train,b_test=train_test_split(a,b,test_size=0.2,random_state=0)
a_train.shape,a_test.shape,b_train.shape,b_test.shape
```

```python
# Heart
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
y_train= np.array(y_train).reshape(-1, 1)
y_test= np.array(y_test).reshape(-1, 1)
```

```python
# Air
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

a_train_scaled = scaler.fit_transform(a_train)
a_test_scaled = scaler.transform(a_test)
b_train= np.array(b_train).reshape(-1, 1)
b_test= np.array(b_test).reshape(-1, 1)
```

```python
# Heart
from sklearn.preprocessing import LabelEncoder

heart_label= ['age','chol']

encoder = LabelEncoder()
for col in heart_label:
        heart[col] = encoder.fit_transform(heart[col])
```

```python
# Air
from sklearn.preprocessing import LabelEncoder

air_label= ['AQI_Bucket']

encoder = LabelEncoder()
for col in air_label:
        air[col] = encoder.fit_transform(air[col])
```

```python
# Data model building
```

```python
# Logistic Regression model
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.linear_model import LogisticRegression

heart_model = LogisticRegression()
heart_model.fit(x_train_scaled, y_train)

y_pred = heart_model.predict(x_test_scaled)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
# Decision tree model
from sklearn.tree import DecisionTreeClassifier

heart_tc=DecisionTreeClassifier(criterion='entropy')
heart_tc.fit(x_train_scaled,y_train)
y_pred=heart_tc.predict(x_test_scaled)

print("Training Accuracy Score :",accuracy_score(y_pred,y_test))
print("Training Confusion Matrix  :",confusion_matrix(y_pred,y_test))
```
