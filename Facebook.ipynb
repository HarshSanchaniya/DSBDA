# Create Data Subsets

```python
# First subset: Like and Share
df_subset_1 = df[['like','share']]
df_subset_1
```

```python
# second subset: Comment and Type
df_subset_2 = df[['comment','Type']]
df_subset_2
```

# Merge Data

```python
merged_data  = pd.merge(df_subset_2, df_subset_1, left_on='comment', right_on= 'like')
merged_data
```

#Example for merge
